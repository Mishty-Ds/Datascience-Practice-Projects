```python
#Model building
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
weather_df = weather_df.apply(LabelEncoder().fit_transform)
```

```python
y = weather_df["Next_Tmax"]
```

```python
z=weather_df["Next_Tmin"]
```

```python
x = weather_df.drop(["Next_Tmax","Next_Tmin"], axis = 1)
```

```python
x
```

```python
z
```

```python
y
```

```python
#deviding data into training and testing sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
x_train,x_test,z_train,z_test=train_test_split(x,z,test_size=0.20,random_state=42)
```

```python
#importing decision tree model 
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy',random_state=0)
classifier.fit(x,y)

classifier_z=DecisionTreeClassifier(criterion='entropy',random_state=0)
classifier_z.fit(x,z)
```

```python
#predicting values for x_test and compairing result with y_test
print("Next_tmax:-")
y_pred=classifier.predict(x_test)
print("predicted values:-")
print(y_pred.astype(int))
y_test_arr=np.array(y_test)
print("original values:-")
print(y_test_arr)

```

```python
#predicting values for x_test and compairing result with z_test
print("Next_Tmin:-")
z_pred=classifier_z.predict(x_test)
print("predicted values:-")
print(z_pred.astype(int))
z_test_arr=np.array(z_test)
print("original values:-")
print(z_test_arr)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#cheaking accuracy of our model
accuracy=accuracy_score(y_pred,y_test)
print("{}%".format(accuracy*100))

accuracy=accuracy_score(z_pred,z_test)
print("{}%".format(accuracy*100))
```

```python
from sklearn.metrics import mean_absolute_error, median_absolute_error
```

```python


# make a prediction set using the test set
prediction = regressor.predict(x_test)

# Evaluate the prediction accuracy of the model 
print('The Explained Variance: %.2f' % regressor.score(x_test, y_test))
print('The Mean Absolute Error: %.2f degrees celcius' % mean_absolute_error(
    y_test, prediction))
print('The Median Absolute Error: %.2f degrees celcius' %
      median_absolute_error(y_test, prediction))
```
