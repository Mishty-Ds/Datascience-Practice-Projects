```python
#Dropping the not so important features 
X = X.drop(['Workclass', 'Education', 'Race', 'Sex',
            'Capital_loss', 'Native_country'], axis=1)
```

```python
#To remove the imbalance scaling the data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
for col in X.columns:
    scaler = StandardScaler()
    X[col] = scaler.fit_transform(X[col].values.reshape(-1, 1))
```

```python
#Data Splitting for Modelling
```

```python
from sklearn.model_selection import train_test_split
```

```python
y=df['Income']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(random_state=42)
```

```python
log_reg.fit(X_train, y_train)
```

```python
Y_pred_logreg = log_reg.predict(X_test)
```

```python
#KNN Classifier
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train, y_train)
```

```python
Y_pred_knn = knn.predict(X_test)
```

```python
#Naive Bayes Classifier
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
```

```python
nb.fit(X_train, y_train)
```

```python
Y_pred_nb=nb.predict(X_test)
```

```python
#Decision Tree classifier
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=42)
```

```python
dtc.fit(X_train, y_train)
```

```python
Y_pred_dtc = dtc.predict(X_test)
```

```python
#Random forest Classifier
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42)
```

```python
rf.fit(X_train, y_train)
```

```python
Y_pred_rf = rf.predict(X_test)
```

```python
#Model evaluation via accuracy score
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Logistic Regression:')
print('Accuracy score:', round(accuracy_score(y_test, Y_pred_logreg) * 100, 2))
```

```python
print('KNN Classifier:')
print('Accuracy score:', round(accuracy_score(y_test, Y_pred_knn) * 100, 2))
```

```python
print('Naive Bayes Classifier:')
print('Accuracy score:', round(accuracy_score(y_test, Y_pred_nb) * 100, 2))
```

```python
print('Decision Tree Classifier:')
print('Accuracy score:', round(accuracy_score(y_test, Y_pred_dtc) * 100, 2))
```

```python
print('Random Forest Classifier:')
print('Accuracy score:', round(accuracy_score(y_test, Y_pred_rf) * 100, 2))
```

```python
# On basis of accuracy score Random forest seems to be performing the best hence Hyperparameter tuning
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators = [int(x) for x in np.linspace(start=40, stop=150, num=15)]
max_depth = [int(x) for x in np.linspace(40, 150, num=15)]
```

```python
param_dist = {
    'n_estimators': n_estimators,
    'max_depth': max_depth,
}
```

```python
rf_tuned = RandomForestClassifier(random_state=42)
```

```python
rf_cv = RandomizedSearchCV(
    estimator=rf_tuned, param_distributions=param_dist, cv=5, random_state=42)
```

```python
rf_cv.fit(X_train, y_train)
```

```python
rf_cv.best_score_
```

```python
rf_cv.best_params_
```

```python
rf_best = RandomForestClassifier(
    max_depth=102, n_estimators=40, random_state=42)
```

```python
rf_best.fit(X_train, y_train)
```

```python
Y_pred_rf_best = rf_best.predict(X_test)
```

```python
print('Random Forest Classifier:')
print('Accuracy score:', round(accuracy_score(y_test, Y_pred_rf_best) * 100, 2))
```

```python
#Confusion Matrix
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, Y_pred_rf_best)
```

```python
sns.heatmap(cm, annot=True, fmt='d', cmap='YlGnBu')
```
