2.Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code

```python
#Lets connect to webdriver
driver=webdriver.Chrome(r"C:/chromedriver")
```

```python
#Opening the homepage
driver.get("https://www.bcci.tv/")
```

```python
#To search for international fixture
button=driver.find_element_by_xpath('//a[@class="navigation__link navigation__link--in-drop-down"]')
```

```python
#Get the hyperlink to go to international fixture page
search=button.get_attribute('href')
```

```python
#Go to the page
driver.get(search)
```

```python
#To find all the Match urls to get the other details from each
Match_urls=[]
for j in driver.find_elements_by_xpath('//div[@class="js-list"]/a'):
    Match_urls.append(j.get_attribute('href'))
```

```python
Match_urls
```

```python
import time
```

```python
#Scrape the details for required attributes 

Match_title=[]
Series=[]
Place=[]
Date=[]
Time=[]
for i in Match_urls:
    driver.get(i)
    time.sleep(2)
    try:
        Match_title.append(driver.find_element_by_xpath('//span[@class="mc-header-info__match-description-match-type mc-theme-color"]').text)
    except NoSuchElementException:
        Match_title.append("-")
        
    try:
        Series.append(driver.find_element_by_xpath('//p[@class="mc-header-info__match-description"]').text)
    except NoSuchElementException:
        Series.append("-")
        
    try:
        Place.append(driver.find_element_by_xpath('//h1[@class="mc-header-info__title"]').text)
        
    except NoSuchElementException:
        Place.append("-")
        
    try:
        Date.append(driver.find_element_by_xpath('//div[@class="mc-header-scorebox__datetime"]/strong').text)
        
    except NoSuchElementException:
        Date.append("-")
        
    try:
        Time.append(driver.find_element_by_xpath('//span[@class="mc-header-scorebox__ist-time"]').text)
        
    except NoSuchElementException:
        Time.append("-")
        
        
```

```python
Match_title
```

```python
Series
```

```python
Place
```

```python
Date
```

```python
Time
```

```python
#to put it in a dataframe
Match_details=pd.DataFrame({})
```

```python
Match_details['Match title']=Match_title
Match_details['Series']=Series
Match_details['Place']=Place
Match_details['Date']=Date
Match_details['Time']=Time
```
